```python
import csv

reader = csv.DictReader(open('RAKE_lc_9_1_filtered.csv', 'r'))
keywords_list = []

for line in reader:
    keywords_list.append(line)

keywords_list
```

```python
#get a dictionary of unique keywords, scores averaged, doc_count--year doesn't matter for now

u_keywords = {}

def avg(curr_score, prev_dc, add_score):
    den = prev_dc + 1
    num = float(curr_score)*prev_dc + float(add_score)
    return num/den

for word in keywords_list:
    if word['keyword'] in u_keywords:
        u_keywords[word['keyword']]['score'] = avg(u_keywords[word['keyword']]['score'], u_keywords[word['keyword']]['doc_count'], word['score'])
        u_keywords[word['keyword']]['doc_count'] += 1
    else:
        u_keywords[word['keyword']] = {
            'doc_count': 1,
            'score': word['score']
        }
        
u_keywords
```

```python
len(u_keywords)
```

```python
#get n-grams (trigram+) and bigrams in two different lists

ngrams = []
bigrams = []

for key in u_keywords.keys():
    if len(key.split()) >= 3:
        ngrams.append(key)

for key in u_keywords.keys():
    if len(key.split()) == 2:
        bigrams.append(key)
        
bigrams
```

```python
#check for n-grams containing bigrams, add to dictionary

bi_to_n = {} #bigram : list[n-grams]

for big in bigrams:
    for ng in ngrams:
        n = ng.split()
        b_in_n = False
        for i in range(len(n)-1):
            tmp = n[i] + " " + n[i+1]
            if tmp == big:
                b_in_n = True
                break
        if b_in_n:
            if big in bi_to_n:
                bi_to_n[big].append(ng)
            else:
                bi_to_n[big] = [ng]
            
bi_to_n
```

```python
#list of dictioinaries: bigram, n-gram (containing bigram), bigram score, n-gram score, bigram doc_count, n-gram doc_count
final = []

for big, lst in bi_to_n.items():
    for ng in lst:
        final.append({
            'bigram': big,
            'n-gram (containing bigram)': ng,
            'bigram score': u_keywords[big]['score'],
            'bigram doc_count': u_keywords[big]['doc_count'],
            'n-gram score': u_keywords[ng]['score'],
            'n-gram doc_count': u_keywords[ng]['doc_count']
        })
        
final
```

```python
with open('bigram_to_ngram.csv', 'w', newline='') as f:
    header_present = False
    for bigram in final:
        if not header_present:
            w = csv.DictWriter(f, bigram.keys())
            w.writeheader()
            header_present = True
        w.writerow(bigram)
```
