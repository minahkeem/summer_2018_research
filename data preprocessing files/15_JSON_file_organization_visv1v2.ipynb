```python
import json
```

```python
with open('final.json') as file:
    data = json.load(file)
    
data
```

```python
#create a simplified data that only contains information about unigrams (and includes total count of 'total_count')

simp_data = []
for uni in data:
    tot_count = 0
    for count in uni['total_count']:
        tot_count += count['count']
    simp_data.append({
        'unigram': uni['word'],
        'total_docs': tot_count,
        'total_count': uni['total_count'],
    })
    
simp_data
```

```python
#sorted simp_data for visualization v1

v1 = sorted(simp_data, key = lambda k: k['total_docs'], reverse = True)
v1
```

```python
with open('vis_v1.json', 'w+') as f:
    json.dump(v1, f, indent=4)
```

```python
#create a list of normalized doc_count data
v2 = []

for uni in v1:
    total_count = []
    for year in uni['total_count']:
        total_count.append({
            'year': year['year'],
            'normalized_count': year['count']/uni['total_docs']
        })
    v2.append({
        'unigram': uni['unigram'],
        'total_docs': uni['total_docs'],
        'proportion': total_count
    })

v2
```

```python
with open('vis_v2.json', 'w+') as f:
    json.dump(v2, f, indent=4)
```
