```python
import csv

reader = csv.DictReader(open('RAKE_lc_9_1_filtered.csv', 'r'))
keywords_list = []

for line in reader:
    keywords_list.append(line)

keywords_list
```

```python
#get a sample set of all keywords (top n keywords with highest scores)

sorted_keywords = sorted(keywords_list, key = lambda k: k['score'], reverse = True)

keywords_set = set()
n=1000
i = 0

while len(keywords_set) < n:
    keywords_set.add(sorted_keywords[i]['keyword'])
    i += 1

lst = list(keywords_set)
lst
```
